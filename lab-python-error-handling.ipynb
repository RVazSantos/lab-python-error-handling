def initialize_inventory(products):

        while not valid_quantity:

                quantity = int(input(f"Enter the quantity of {product}s available: "))

````python
#previous data
products = ["t-shirt", "mug", "hat", "book", "keychain"]

#step 1 from lab comprehension + ajusted error handling

def initialize_inventory(products):
    inventory = {}
    for product in products:
        while True:
            try:
                quantity = int(input(f"Enter the quantity of {product} available: "))
                if quantity < 0:
                    raise ValueError("Quantity cannot be negative.You need to repeat the input.")
                inventory[product] = quantity
                break
            except ValueError as error:
                print(f"Error: {error}")
    return inventory



#step 2  from lab comprehension
def get_customer_orders():
    customer_orders = set()
    while True:
        try:
            orders = int(input("Enter the number of different products the customer wants to order: "))
            if orders < 0:
                raise ValueError("Number of orders cannot be negative.")
            break
        except ValueError as error:
            print(f"Error: {error}")
            continue
    for i in range(orders):
        while True:            
            try:
                p = input("Enter the name of a product that a customer wants to order:")
                if p not in inventory:
                    raise ValueError("Product not found in inventory. Please enter a valid product name.")
                if p in customer_orders:
                    raise ValueError("Product already ordered. Please enter a different product.")
                customer_orders.add(p)
                break
            except ValueError as error:
                print(f"Error: {error}")
    return customer_orders


#step 3  from lab comprehension
def update_inventory(customer_orders, inventory):
    inventory_updated = {product: inventory.get(product) - 1 if product in customer_orders else inventory.get(product) for product in inventory}
    inventory_updated = {product: inventory_updated[product] for product in inventory_updated if inventory_updated[product] > 0}
    return inventory_updated

#step 4 from lab comprehension
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_ordered = round((total_products_ordered / len(products)) * 100, 1)
    order_statistics = (total_products_ordered, percentage_ordered)
    return order_statistics

#step 5 from lab comprehension
def print_order_statistics(order_statistics):
    print("Order Statistics:")
    print(f"Total products ordered: {order_statistics[0]}")
    print(f"Percentage of products ordered: {order_statistics[1]:.2f}%")
    return order_statistics

#step 6 from lab comprehension
def print_updated_inventory(inventory):
    print("Updated Inventory:")
    {print(f"{product}: {inventory[product]}") for product in inventory}

#new step 3 from lab comprehension
def calculate_total_price(customer_orders):
    prices = {}
    for product in customer_orders:
        while True:
            try:
                price = float(input(f"Enter the price for {product}: "))
                if price < 0:
                    raise ValueError("Price cannot be negative.")
                prices[product] = price
                break
            except ValueError as error:
                print(f"Error: {error}")
    total_price = sum([prices[product] for product in prices])
    print(f"Total price: {total_price:.1f}")
    return total_price

# Main program from lab comprehension

inventory = initialize_inventory(products)#to set up quantities
customer_orders = get_customer_orders()#to record the order from the customer
inventory = update_inventory(customer_orders, inventory)#to update the inventory after the order
order_statistics = calculate_order_statistics(customer_orders, products)#calculate statistics
print_order_statistics(order_statistics)#to print statistics
print_updated_inventory(inventory)#to print the updated inventory
calculate_total_price(customer_orders)#to calculate total price of the order
````
